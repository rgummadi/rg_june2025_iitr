```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# 10,000 random values from a normal distribution
x = np.random.normal(loc=0, scale=1, size=10000)

mean_x = x.mean()
median_x = np.median(x)

# Density curve
pd.Series(x).plot(kind="density", figsize=(8, 5), label="Density")

# Mean and median lines
plt.axvline(mean_x, color="red", linestyle="--", linewidth=2, label=f"Mean = {mean_x:.3f}")
plt.axvline(median_x, color="green", linestyle="--", linewidth=2, label=f"Median = {median_x:.3f}")

plt.title("Density Curve of 10,000 Normally Distributed Numbers")
plt.xlabel("Value")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Exponentially distributed values
x = np.random.exponential(size=100000)

mean_x = x.mean()
median_x = np.median(x)

# Density curve
pd.Series(x).plot(kind="density", figsize=(8, 5), label="Density")

# Mean and median lines
plt.axvline(mean_x, color="red", linestyle="--", linewidth=2, label=f"Mean = {mean_x:.3f}")
plt.axvline(median_x, color="green", linestyle="--", linewidth=2, label=f"Median = {median_x:.3f}")

plt.title("Density Curve of Exponentially Distributed Numbers")
plt.xlabel("Value")
plt.legend()
plt.show()
```

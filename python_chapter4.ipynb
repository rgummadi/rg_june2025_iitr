
**Exercise:**
Write a function `compute_mean(numbers)` that takes a list of numbers and returns their mean (average).
If the list is empty, return `0`.

```python
def compute_mean(numbers):
    arraylen = len(numbers)
    if (arraylen == 0):
        return 0
    sum, count = 0,0
    while  count < arraylen:
        sum += numbers[count]
        count += 1
    return sum/arraylen

print(compute_mean([2, 4, 6, 8]))      # Output: 5.0
print(compute_mean([10, 20, 30]))      # Output: 20.0
print(compute_mean([]))               # Output: 0
```


**Exercise:**
Write a function `compute_sd(numbers)` that takes a list of numbers and returns the standard deviation.

```python
def compute_sd(numbers):
    mean = compute_mean(numbers)
    squared_diff = []
    for number in numbers:
        squared_diff.append((number - mean)**2)

    avg_sd = compute_mean(squared_diff)
    return (avg_sd ** 1/2)


print(compute_sd([2, 4, 4, 4, 5, 5, 7, 9]))   # Output: 2.0
print(compute_sd([10, 10, 10, 10]))           # Output: 0.0
```

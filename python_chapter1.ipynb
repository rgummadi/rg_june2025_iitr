```python

## Calculate Interest

def calculate_interest(principal, rate, time):
    return principal * (rate / 100) * time


print (calculate_interest(1000, 5, 2))   
print (calculate_interest(1500, 4.3, 3))
print(calculate_interest(500, 10, 0))
```

```python
## Calculate Hypotense

def calculate_hypotenuse(base, height):
    return (base**2 + height**2)**0.5

print(calculate_hypotenuse(3, 4))
print(calculate_hypotenuse(5, 12))
```

```python

## Find Distance 2D
import math
def find_distance_2d(x1, y1, x2, y2):
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

print(find_distance_2d(0, 0, 3, 4))     
print(find_distance_2d(1, 2, 4, 6))     
```

```python

## Find Distance 3D
import math
def calculate_distance_3d(x1, y1, z1, x2, y2, z2):
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2 + (z2 - z1)**2)

print(calculate_distance_3d(0, 0, 0, 1, 1, 1))  # Output: 1.7320508075688772
print(calculate_distance_3d(2, 3, 5, 2, 3, 5))  # Output: 0.0
print(calculate_distance_3d(1, 2, 3, 4, 6, 9))  # Output: 7.810249675906654
```

```python
#manhattan distance
def manhattan_distance(x1, y1, x2, y2):
    return abs(x2 - x1) + abs(y2 - y1)


print(manhattan_distance(2, 3, 5, 1))  # Output: 5
print(manhattan_distance(0, 0, 0, 0))  # Output: 0
print(manhattan_distance(-1, -1, 1, 1))  # Output: 4
```

```python

#Predict Y on a Straight Line (2D)
def predict(m,c,x):
    return m * x + c
print(predict(2, 3, 4))     # Output: 11
print(predict(-1, 5, 2))    # Output: 3
```

```python
#Write a function `fit(x1, y1, x2, y2)` that returns a tuple `(m, c)`
 # representing the slope and intercept of the line passing through the two points `(x1, y1)` and `(x2, y2)`.

def fit(x1, y1, x2, y2):
    if x2 - x1 == 0:
        raise ValueError("x1 and x2 cannot be the same value (vertical line)")
    m = (y2 - y1) / (x2 - x1)
    c = y1 - m * x1
    return (m, c)

print(fit(1, 2, 3, 6))       # Output: (2.0, 0.0)
print(fit(0, 5, 2, 9))       # Output: (2.0, 5.0)
```

```python
#First derivative of a function
import math
def approx_derivative(f, x, h):
    return (f(x +h) - f(x)) / h

def square(x):
    return x * x

def cube(x):
    return x * x * x

def sine(x):
    return math.sin(x)

def exp(x):
    return math.exp(x)

print(approx_derivative(square, 3, 0.0001))  
print(approx_derivative(cube, 2, 0.0001))    
print(approx_derivative(sine, 2, 0.0001))    
print(approx_derivative(exp, 2, 0.0001))    
```

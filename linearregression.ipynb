#Gradient Descent

```python
import numpy as np
x1 = np.random.random(400)
x2 = np.random.random(400)
noise = (np.random.random(400) - 0.5)/4
y = 3*x1 + 5*x2 + 10 + noise

```

```python
def error(m1,m2,c):
    ypredict = m1*x1 + m2*x2 + c
    errors = ypredict - y
    return np.mean(errors**2)
```

```python
def diff2(fn,idx,*args):
    delta = 0.000000000001
    y = fn(*args)
    args = list(args)
    args[idx] += delta
    y1 = fn(*args)
    return (y1-y)/delta
```

```python
m1,m2,c = np.random.random(3,)
epoch = 2000
for i in range(epoch):
    eta = 0.01
    gradm1 = diff2(error,0,m1,m2,c)
    gradm2 = diff2(error,1,m1,m2,c)
    gradc = diff2(error,2,m1,m2,c)
    m1 = m1 - eta * gradm1
    m2 = m2 - eta * gradm2
    c = c - eta * gradc
    e = error(m1,m2,c)
    print(m1,m2,c,e)
```
